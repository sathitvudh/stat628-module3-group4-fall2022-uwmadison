```python
import csv
import json
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
import string
import pickle
```

### Data tokenization and cleaning

```python
df = pd.concat(map(pd.read_csv, ['./data/bars_review_1.csv', './data/bars_review_2.csv', './data/bars_review_3.csv', './data/bars_review_4.csv', './data/bars_review_5.csv']), ignore_index=True).drop(columns=['Unnamed: 0'])
```

```python
review_word = [nltk.word_tokenize(review.lower().translate(str.maketrans('', '', string.punctuation))) for review in df['text']]
```

```python
with open("review_word_list", "wb") as f:
    pickle.dump(review_word, f)
```

```python
with open("review_word_list", "rb") as f:
    test = pickle.load(f)
```

```python
# remove punctuations
punctuation = string.punctuation
for review in test:
    review.remove(list(string.punctuation))
```
