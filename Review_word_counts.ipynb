```python
import csv
import json
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer 
import string
import pickle
```

```python
bars_review = pd.read_csv("./data/bars_list_ca.csv").drop(columns=['Unnamed: 0'])
```

```python
bars_review
```

```python
# Tokenize, lowercase, remove punctuations

review_word = [nltk.word_tokenize(review.lower().translate(str.maketrans('', '', string.punctuation))) for review in bars_review['text']]
```

```python
# Stemming word (Snowball Stemmer)

sb_stemmer = SnowballStemmer("english",)
review_stem_word = []
for review in review_word:
    for word in review:
        stemmer = sb_stemmer.stem(word)
        review_stem_word.append(stemmer)
```
