### Step 1: clean and filter data

```python
categories = set()

for i in range(len(business)):
    if business[i]['categories'] is None:
        continue # skip
        
    else:
        cat = set(business[i]['categories'].split(", "))
        categories = categories.union(cat)
```

```python
business_dict = {}

for i in range(len(business)):
    business_dict[business[i]['business_id']] = [business[i]['name'], business[i]['categories']]
```

```python
bars_list = {}

for i in business_dict:
    #print(i)
    #search_word = ["Gay Bars","Cocktail Bars","Dive Bars","Wine Bars",
    #               "Whiskey Bars","Bars","Champagne Bars","Beach Bars",
    #               "Sports Bars","Tapas Bars","Piano Bars","Vermouth Bars",
    #               "Beer Bar","Tiki Bars"]
    search_word = ["Cocktail Bars","Wine Bars","Whiskey Bars","Champagne Bars","Beer Bar"]
    for key in search_word:
        if re.search(key, str(business_dict.get(i))):
            #print(i, "--", business_dict.get(i))
            bars_list[i] = business_dict.get(i)[0]
```
